```python
import pandas as pd
import os  
import glob 
```

```python
folder_path = "./savivaldybes/"
df_list = []

for file in glob.glob(os.path.join(folder_path, "*_a_z.json")): 

    savivaldybe = os.path.basename(file).replace('_a_z.json', '')

    temp_df = pd.read_json(file) 
    temp_df['savivaldybe'] = savivaldybe 

    df_list.append(temp_df)  

final_df = pd.concat(df_list, ignore_index=True)  
final_df = final_df.drop(columns=["url"])
```

```python
final_df.to_csv("sav.csv", index=False)
```
